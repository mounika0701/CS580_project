# **Problem 1**

```python
# Problem 1:
# Creating R1 and R2
R1 = [(47, 7), (67, 54), (10, 20), (61, 10), (91, 32), (96, 5), (24, 65), (20, 64), (32, 51), (65, 42)]
R2 = [(7, 12), (34, 54), (10, 20), (61, 10), (10, 32), (101, 202), (24, 65), (5, 81), (32, 51), (42, 1)]

# Creating a hash table for R2 with attribute B as key
hash_table, res = {}, []

# Looping through each row of R2 to build the hash table
for rowR2 in R2:
    key = rowR2[0]  # Value of attribute B
    if key in hash_table:
        hash_table[key].append(rowR2)
    else:
        hash_table[key] = [rowR2]

# Evaluating the query q(A, B, C) :−R1(A, B), R2(B, C)
# Looping through each row of R1 to find matches in R2 based on attribute B
for rowR1 in R1:
    key = rowR1[1]  # Value of attribute B
    if key in hash_table:
        # If a match is found in R2, append the result tuple to 'res'
        for rowR2 in hash_table[key]:
            res.append((rowR1[0], rowR1[1], rowR2[1]))

# Printing the result
print("R1 = ", R1)
print("R2 = ", R2)
print("Result of the query q(A, B, C) : −R1(A, B), R2(B, C):")
for tuple_result in res:
    print(tuple_result)
```

# **Problem 2**

```python
# Problem 2
from collections import defaultdict

def hash_join(Ra, Rb):
    hash_map = defaultdict(list)
    for b in Rb:
        hash_map[b[0]].append(b)
    result = []
    for a in Ra:
        if a[1] in hash_map:
            for b in hash_map[a[1]]:
                result.append(a + b[1:])
    return result

def evaluate_k_line_join_query(query, relations):
    S = None
    for relation in query:
        projection = [(tuple_[i], tuple_[i+1]) for tuple_ in relations[relation] for i in range(len(tuple_)-1)]
        if S is None:
            S = projection
        else:
            S = hash_join(S, projection)
    return S

# Example usage
query = ["R1", "R2", "R3"]
relations = {
    "R1": [(1, 2), (2, 3), (3, 4)],
    "R2": [(2, 3), (3, 4), (4, 5)],
    "R3": [(3, 4), (4, 5), (5, 6)]
}

result = evaluate_k_line_join_query(query, relations)
print("Result of k-line join query:", result)
print("Number of tuples in the result:", len(result))
```

# **Problem 3**

```python
import time

class HashJoin:
    def __init__(self, tuples):
        # Initializing a hash table
        self.hash_table = {}
        for i in tuples:
            key = i[-1]
            if key in self.hash_table:
                self.hash_table[key].append(i)
            else:
                self.hash_table[key] = [i]
    def __iter__(self):
        # Iterator over the hash table
        for key in self.hash_table:
            for tuple in self.hash_table[key]:
                yield tuple

    def join(self, other):
        # Performing a hash join with another HashJoin instance
        res = []
        for i in other:
            # The join condition involves the first attribute
            key = i[0]
            if key in self.hash_table:
                for same_tuple in self.hash_table[key]:
                    # Joining matching tuples from both relations
                    append_q = list(same_tuple) + list(i[1:])
                    res.append(append_q)
        return res

if __name__ == "__main__":
    # Creating instances of HashJoin for each relation
    R1 = HashJoin([(1, 'a'), (2, 'b'), (3, 'c'),(4, 'd'), (5, 'e'), (6, 'f'),(7, 'g'), (8, 'h'), (9, 'i'), (10, 'j')])
    R2 = HashJoin([('a', 'p'), ('b', 'q'), ('c', 'r'),('d', 's'), ('e', 't'), ('f', 'u'),('g', 'v'), ('h', 'w'), ('i', 'x'),('j', 'y')])
    R3 = HashJoin([('p', 10), ('q', 20), ('r', 30),('s', 40), ('t', 50), ('u', 60),('v', 70), ('w', 80), ('x', 90),('y', 100)])
    R4 = HashJoin([(10, 1.1), (20, 2.2), (30, 3.3),(40, 4.4), (50, 5.5), (60, 6.6),(70, 7.7), (80, 8.8), (90, 9.9),(10, 10.10)])

    start = time.time()
    result = R1.join(R2)
    result = HashJoin(result).join(R3)
    result = HashJoin(result).join(R4)
    # Printing the results
    print("Result of hash join query:")
    for line in result:
        print(line)
    end = time.time()
    print("\nTime Taken = {:.3f}s".format(end - start))
```

```python
import random
import time

random.seed(0)  # Ensure consistent data generation

# Generating dataset for R1, R2, and R3
R1 = [(i, random.randint(1, 5000)) for i in range(1, 101)]
R2 = [(random.randint(1, 5000), j) for j in range(1, 101)]
R3 = [(l, l) for l in range(1, 101)]

def yannakakis_algorithm(R1, R2, R3):
    start_time = time.time()
    join1 = {b: [] for a, b in R1}
    for a, b in R1:
        join1[b].append((a, b))
    intermediate_result = []
    for b, c in R2:
        if b in join1:
            for a, _ in join1[b]:
                intermediate_result.append((a, b, c))
    join2 = {c: [] for _, _, c in intermediate_result}
    for a, b, c in intermediate_result:
        join2[c].append((a, b, c))
    final_result = []
    for c, d in R3:
        if c in join2:
            for a, b, _ in join2[c]:
                final_result.append((a, b, c, d))
    execution_time = time.time() - start_time
    return final_result, execution_time

def hash_join(R1, R2, R3):
    start = time.time()
    R2_map = {}
    for b, c in R2:
        R2_map.setdefault(b, []).append(c)
    R1_R2_join = []
    for a, b in R1:
        if b in R2_map:
            for c in R2_map[b]:
                R1_R2_join.append((a, b, c))
    R3_map = {}
    for c, d in R3:
        R3_map.setdefault(c, []).append(d)
    final_result = []
    for a, b, c in R1_R2_join:
        if c in R3_map:
            for d in R3_map[c]:
                final_result.append((a, b, c, d))
    final_time = time.time() - start
    return final_result, final_time

result_yannakakis, time_yannakakis = yannakakis_algorithm(R1, R2, R3)
result_hash, time_hash = hash_join(R1, R2, R3)
result_yannakakis.sort()
result_hash.sort()

print("Yannakakis Algorithm Results:")
print(f"Number of tuples: {len(result_yannakakis)}")
print(f"Execution Time: {time_yannakakis:.6f} seconds")
print(f"Result of Yannakakis\n", result_yannakakis)

print("\nHash Join Results:")
print(f"Number of tuples: {len(result_hash)}")
print(f"Execution Time: {time_hash:.6f} seconds")
print(f"Result of Hash Join:\n", result_hash)
```

```python
import random
import time

# Creating the dataset for Problem 5
R1 = [(i, 5) for i in range(1, 1001)] + [(i, 7) for i in range(1001, 2001)]
R1.append((2001, 2002))
random.shuffle(R1)

R2 = [(5, i) for i in range(1, 1001)] + [(7, i) for i in range(1001, 2001)]
R2.append((2002, 8))
random.shuffle(R2)

R3 = [(random.randint(2002, 3000), random.randint(1, 3000)) for _ in range(2000)]
R3.append((8, 30))
random.shuffle(R3)

# Implementing Yannakakis Algorithm for Problem 2
def yannakakis_algorithm(R1, R2, R3):
    start_time = time.time()
    # Join R1 and R2 on the second attribute of R1 and first of R2
    join1 = {b: [] for a, b in R1}
    for a, b in R1:
        join1[b].append((a, b))

    intermediate_result = []
    for b, c in R2:
        if b in join1:
            for a, _ in join1[b]:
                intermediate_result.append((a, b, c))

    # Join intermediate_result with R3
    join2 = {c: [] for _, _, c in intermediate_result}
    for a, b, c in intermediate_result:
        join2[c].append((a, b, c))

    final_result = []
    for c, d in R3:
        if c in join2:
            for a, b, _ in join2[c]:
                final_result.append((a, b, c, d))

    execution_time = time.time() - start_time
    return final_result, execution_time

# Implementing Sequential Join for Problem 3
def hash_join(R1, R2, R3):
    start = time.time()
    # Creating a hash map for R2
    R2_map = {}
    for b, c in R2:
        if b in R2_map:
            R2_map[b].append(c)
        else:
            R2_map[b] = [c]

    # Join R1 with R2 using the hash map
    R1_R2_join = []
    for a, b in R1:
        if b in R2_map:
            for c in R2_map[b]:
                R1_R2_join.append((a, b, c))

    # Creating a hash map for R3
    R3_map = {}
    for c, d in R3:
        if c in R3_map:
            R3_map[c].append(d)
        else:
            R3_map[c] = [d]

    # Final join of R1_R2_join with R3 using the hash map
    res = []
    for a, b, c in R1_R2_join:
        if c in R3_map:
            for d in R3_map[c]:
                res.append((a, b, c, d))

    final_time = time.time() - start
    return res, final_time


# Running both implementations
result_yannakakis, time_yannakakis = yannakakis_algorithm(R1, R2, R3)
result_hash, time_hash = hash_join(R1, R2, R3)

print("Yannakakis Algorithm Results:")
print(f"Number of tuples: {len(result_yannakakis)}")
print(f"Execution Time: {time_yannakakis:.4f} seconds")
result_yannakakis.sort()
print(f"Result of Yannakakis\n", result_yannakakis)

print("\nHash Join Results:")
print(f"Number of tuples: {len(result_hash)}")
print(f"Execution Time: {time_hash:.4f} seconds")
result_hash.sort()
print(f"Result of Hash Join:\n", result_hash)
```
